```python
import requests
from bs4 import BeautifulSoup


# Send a GET request to the website
url = "https://www.pbs.gov.pk/content/final-results-census-2017-0"
response = requests.get(url)

# Use BeautifulSoup to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")



# find all the links on the page
links = soup.find_all("a")


```

```python
links
```

```python
desired_links = []

for link in links:
    if 'a href="/sites/default/files' in str(link):
        desired_links.append(link)
```

```python
desired_links
```

```python
count = len(desired_links)
print("Number of desired links:", count)
```

```python
for link in desired_links:
    href = link.get('href')
    print(href)
```

```python
hrefs = []
prefix = 'https://www.pbs.gov.pk'

for link in desired_links:
    href = link.get('href')
    href_with_prefix = prefix + href
    hrefs.append(href_with_prefix)

hrefs
```

```python
count = len(hrefs)
print("Number of desired links:", count)
```

```python
#downloading the files

import os
# create a new directory to save downloaded files
folder_path = "tables_files"
if not os.path.exists(folder_path):
    os.mkdir(folder_path)

# loop over the hrefs and download each file
for href in hrefs:
    # extract the state/province name and table number from the href
    split_href = href.split("/")
    state_province = split_href[-2]
    table_number = split_href[-1].split(".")[0]
    
    # name the file
    filename = f"{state_province}_{table_number}.pdf"
    
    # download the file
    response = requests.get(href)
    
    # save the file to the new directory
    with open(os.path.join(folder_path, filename), "wb") as f:
        f.write(response.content)
```

```python
#please write python code to remove the 1 character before .pdf character from the names of pdf files



folder_path = "tables_files"

pdf_files = [f for f in os.listdir(folder_path) if f.endswith(".pdf")]

for filename in pdf_files:
    old_path = os.path.join(folder_path, filename)
    # remove the first character before .pdf
    new_filename = filename[:-5] + ".pdf"
    # make sure the new filename is unique
    count = 0
    while os.path.exists(os.path.join(folder_path, new_filename)):
        count += 1
        new_filename = filename[:-5] + str(count) + ".pdf"
    new_path = os.path.join(folder_path, new_filename)
    # rename the file
    os.rename(old_path, new_path)



```
